## Download S3 files into Ubuntu

```python
import os

base_dir = 'Dissertation'
os.makedirs(base_dir, exist_ok=True)

for prefix in folders:
    print(f"\nStarting download from S3 prefix: {prefix}")
    paginator = s3.get_paginator('list_objects_v2')
    page_iterator = paginator.paginate(Bucket=bucket_name, Prefix=prefix)

    count = 0
    for page in page_iterator:
        contents = page.get('Contents', [])
        
        for obj in contents:
            key = obj['Key']  # e.g. 'haze/image001.jpg'
            local_path = os.path.join(base_dir, key)  # e.g. 'Dissertation/haze/image001.jpg'

            # Ensure the directory exists
            os.makedirs(os.path.dirname(local_path), exist_ok=True)

            # Download the file
            s3.download_file(bucket_name, key, local_path)
            count += 1

    print(f"Finished downloading {count} files from: {prefix}")

print("\nAll folders downloaded successfully!")

```

```python
import os
import boto3

# Assuming s3 client and other variables are already defined:
# s3 = boto3.client('s3')
# bucket_name = 'your-bucket-name'
# folders = ['haze/', 'nonhaze/', ...]  # your prefixes

base_dir = 'Dissertation'
os.makedirs(base_dir, exist_ok=True)
s3 = boto3.client('s3')
bucket_name = 'dic-haze-dataset'
folders = ['haze/', 'nonhaze/','unseen/'] 

for prefix in folders:
    print(f"\nStarting download from S3 prefix: {prefix}")
    paginator = s3.get_paginator('list_objects_v2')
    page_iterator = paginator.paginate(Bucket=bucket_name, Prefix=prefix)

    count_downloaded = 0
    count_skipped = 0

    for page in page_iterator:
        contents = page.get('Contents', [])
        
        for obj in contents:
            key = obj['Key']  # e.g. 'haze/image001.jpg'
            local_path = os.path.join(base_dir, key)  # e.g. 'Dissertation/haze/image001.jpg'

            # Check if file already exists
            if os.path.exists(local_path):
                count_skipped += 1
                continue  # skip downloading this file

            # Ensure the directory exists
            os.makedirs(os.path.dirname(local_path), exist_ok=True)

            # Download the file
            s3.download_file(bucket_name, key, local_path)
            count_downloaded += 1

    print(f"Downloaded {count_downloaded} new files from: {prefix}")
    print(f"Skipped {count_skipped} existing files from: {prefix}")

print("\nAll folders processed successfully!")
```
