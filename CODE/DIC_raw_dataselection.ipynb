```python
import os
import glob
import shutil
import random

# CONFIGURATION
source_base = "D:\\Hari\\2025_06_26"
output_base = "D:\\Dissertation\\unseen"    
os.makedirs(output_base, exist_ok=True)

# Folder names for haze (label=1)
haze_folders = ['0cm', '20cm', '40cm', '60cm', '80cm', '100cm', '120cm']
nonhaze_folder = 'no_heating'  # label=0

# Create output folders
haze_out = os.path.join(output_base, "haze")
nonhaze_out = os.path.join(output_base, "nonhaze")
os.makedirs(haze_out, exist_ok=True)
os.makedirs(nonhaze_out, exist_ok=True)

# HAZE IMAGE SAMPLING
def sample_haze_images():
    total_selected = 0
    for folder in haze_folders:
        path = os.path.join(source_base, folder)
        all_images = sorted(glob.glob(os.path.join(path, "*.tiff")))

        selected = []
        i = 0
        while len(selected) < 28 and i < len(all_images):
            selected.append(all_images[i])
            i += random.randint(8, 12)  # ~every 10th frame with variation

        print(f"Selected {len(selected)} from {folder}")
        for j, img_path in enumerate(selected):
            fname = os.path.basename(img_path)
            newname = f"{folder}_{j:04d}.tiff"
            shutil.copy(img_path, os.path.join(haze_out, newname))
            total_selected += 1

    print(f"Total haze images selected: {total_selected}")

# NON-HAZE IMAGE SAMPLING
def sample_nonhaze_images():
    path = os.path.join(source_base, nonhaze_folder)
    all_images = sorted(glob.glob(os.path.join(path, "*.tiff")))
    step = len(all_images) // 200
    selected = all_images[::step][:200]

    print(f"Selected {len(selected)} non-haze images")
    for j, img_path in enumerate(selected):
        fname = os.path.basename(img_path)
        newname = f"noheat_{j:04d}.tiff"
        shutil.copy(img_path, os.path.join(nonhaze_out, newname))

    print("Non-haze sampling complete.")


sample_haze_images()
sample_nonhaze_images()
print(" Done. Images copied and renamed into 'haze/' and 'nonhaze/' folders.")
```
